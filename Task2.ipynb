Task 2 – Logic and Circuits

1. I chose microwave system to apply logical reasoning to it.

2. Now, I'm going to define logical condition for this system:

A = Start button is pressed (required)

B = Door is closed (required)

C = Microwave has power supply (required)

D = Food is inside (optional; does not affect the system)

So, if we call the output M the microwave will work if and only if A,B and C steps are performed at the same time.
If A,B and C steps are performed but not D, the output would still be the same (1 - true) because D step doesn't affect the system.

3. Now, I'm going to create the truth table for my microwave system:

![Truth Table](truth_table_microwave.png)

1 means true and 0 means false.

As we see, there's only 2 cases out of 16 when the system works.

Boolean equation:

M= A ⋅ B ⋅ C 

Simplified form using De morgan's law:

1. ⋅ = AND

2.+ = OR

3. ! = Negation

M = A ⋅ B ⋅ C

!M = !(A ⋅ B ⋅ C)=!A+!B+!C

As we know, !!M=M, so:  A ⋅ B ⋅ C = !(!A+!B+!C) (using double negation to get the same).

Finally, we get M=!(!A+!B+!C)

5. Purely logical systems are limited because they can only follow the exact inputs and strict rules which are predefined. If anything doesn't fit those rules, that kind of system will fail. They are not designed to handle unclear or messy situations. But in real world, situations often involve uncertainty and things are unpredictable, so pure logic is not always enough.
Machines reason in a consistent way, they are precise with inputs and outputs, and the results are always the same. Humans mostly rely on emotions, intuition and past experience, so even if we aren't always precise, we can still make reasonable decisions without having all the information. Because of this, humans are more flexible, while machines are more fast and accurate.
