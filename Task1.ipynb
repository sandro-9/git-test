1. I will use the word "accommodation" to perform Huffman encoding. I chose that word because it has repeated and also unique letters 
which will help me to demonstrate the efficiency of Huffman encoding better.

2. First, I will calculate the frequency of each character in this word:
a-2 ; c-2 ; o-3 ; m-2 ; d-1 ; t-1 ; i-1 ; n-1.

3. Now, I will build the Huffman tree based on these frequencies step by step:
    - Combine the two least frequent characters d(1) and t(1) to create a node with frequency 2.
    - Combine i(1) and n(1) to create a node with frequency 2.
    - Combine a(2) and c(2) to create a node with frequency 4.
    - Combine m(2) and the node with frequency 2 (d,t) to create a node with frequency 4.
    - Combine o(3) and the node with frequency 2 (i,n) to create a node with frequency 5.
    - Combine the two nodes with frequency 4 (a,c) and 4 (m,d,t) to create a node with frequency 8.
    - Finally, combine the two nodes with frequencies 5 (i,n,o) and 8 (a,c,m,d,t) to create the root node with frequency 13.

![Huffman tree](huffman_tree_accommodation.png)

The letters that appear more often, like o, are higher up and get shorter codes, while the rare ones like d and t are lower in the tree and get longer codes.







4. Now, I will encode my word using the Huffman tree

a - 100;
c - 101;
o - 01;
m - 110;
d - 1110;
t - 1111;
i - 000; 
n - 001;

So, the encoded representation of "accommodation" is: 100 101 101 01 110 110 01 1110 100 1111 000 01 001
it consists of 38 bits in total.

If I encoded that word using standart 8-bit format, it would be: 01100001 01100011 01100011 01101111 01101101 01101101 01101111 01100100 01100001
01110100 01101001 01101111 01101110
which is much longer than the Huffman encoded version and consists of 104 bits in total. 
Using Huffman encoding system, I was able to reduce the size of the representation significantly by using variable-length codes based on character frequencies.

5. We saved 66 bits by using Huffman encoding system instead of standard 8-bit encoding for the word "accommodation", which is a significant reduction in size.
    This shows well the efficiency of using Huffman encoding for texts especially with varying character frequencies.
    In this type of encoding system the structure of the message is important because the idea of Huffman encoding is to allign shorter codes
    with more frequent characters and accordingly longer codes with less frequent ones. Also having the same Huffman tree is important 
    for reciever to decode the message correctly, because in a different Huffman tree the same code can be used for another character. Also I should mention that
    Huffman encoding is a lossless compression method, meaning that no information is lost during the encoding and decoding process, and it is also a prefix-free code which means that no code is a prefix of any other code in the tree which allows for unambiguous decoding of the encoded data.
    Huffman coding is actually used in real life too, for example, in ZIP files, JPEG photos, and MP3 music.
    It helps make files smaller without losing any information, which is exactly what makes it so useful and efficient.
