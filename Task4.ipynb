Redundancy and information loss in digital systems

1.
 In digital systems, ensuring correct data transmission is important. This data consists of sequence of bits ( 0s and 1s ), and if a single bit flips or changes, we came across wrong output. To avoid this outcome, in almost every electronic devices, we use error detection and correction techniques.
 Bits can be flipped or changed due to external factors, such as: noise, hardware glitches and other physical imperfections. After that, the input data is not the same as the output, and this type of mismatch is called “Error”.
 Those errors can cause loss of secured and important information, even though we are just talking about single bit change, so detecting and correcting them are crucial.
 There are generally 3 different types of errors:
1. Single bit errors (The change of just one bit)
2. Multiple bit errors (The change of two or more bits)
3. Burst errors (The change of sets of bits)
 Error detection and error correction are different but interconnected processes. While both of them are crucial for ensuring that codes are reliable and functioning as expected, they still do different things. Error detection also can be automated by machine, while error correction often involves human flexibility and problem-solving skills.
 Error detection is just simply letting the system know that something went wrong. There some ways to detect errors and I will talk about them.
 On the other hand, error correction is not just the process of identifying the error, but locating it precisely and making necessary changes to fix it. It’s more involved process than just detecting errors.
 Firstly, easiest way to detect errors is by adding a parity bit. We just simply add a parity bit to the data before transmitting. The parity bit ensures that the count of 1s in data is even or odd, so there are two types of parity bits: 
 Even parity bit: With even parity, you want the total number of 1s in the bit sequence to be even. So you first count how many 1s you have.
 If the number of 1s is odd, you set the parity bit to 1 to make the total even.
 If the number of 1s is already even, the parity bit becomes 0 because nothing needs to be fixed.

 Odd parity bit: Odd parity works the same way but with the opposite goal, you want the total number of 1s to be odd.
 If the count of 1s is even, you set the parity bit to 1 to make the total odd.
 If it’s already odd, then the parity bit is 0, since the sequence already meets the condition.

 For example, in the case of even parity, the system ensures that total number of 1s in data sequence is even. If the receiver counts and gets odd number of 1s, it means that the data was corrupted.

 Now for the error correction, classic method is the Hamming code. It can locate the error and fix it. It adds multiple redundancy bits in specific positions so if a bit flips, parity bits pattern errors can reveal which exact bit is flipped. For example, Hamming code (7,4) adds 3 parity bits to 4 data, so if only one bit flips, system can detect where it is and correct it automatically. The fact is that Hamming codes add limited redundancy to the data, so it mostly works when error rate is low, for example, computer memory ( mostly RAM). 

 CRC (Cyclic Redundancy Check) is just a method to detect if something went wrong during digital data transmission and storage. Before the data is sent or saved, the system does a quick math on it and calculates a small number (the CRC). That number represents the data, and then both the data and number go together. When the data is checked again later, the system repeats the same math again. If the new number matches the old one then everything’s correct, but if it doesn’t match it means that something got corrupted.

Example of parity bits:

Even parity example:

Max wants to transmit:  10010
Max calculates parity bit value: 1^0^0^1^0 = 0 (because sum of the 1s is even)
Max adds parity bit and sends: 100100

TRANSMISSION ERROR (second bit flips during transmission)

Jack receives: 110100
Jack calculates overall parity: 1^1^0^1^0^0 = 1
Jack notices that there is transmission error because the total count of 1s is odd.

For example, with Hamming codes, several parity bits are placed in positions 1, 2, 4, … and the combination of which parity checks fail tells you exactly which bit is wrong.

3.
Error detection and correction are important techniques because they ensure reliable transmission of the data. Even when we send a message through the internet, there is a chance that it corrupts. We strongly rely on data and even a single bit error can be a big deal, for example, in medicine, incorrect reading in a medical scan, little changes to payment informations and so on. Things like parity bits, hamming code, CRC and so on help the systems to detect errors and sometimes even fix them.

These methods are also used in storage devices. SSDs, hard drives and RAM use different forms of error correction codes to avoid changes or damages in files. Without correcting them, even a single flipped bit could ruin the documents or programs.
On the other hand, there’s always downsides with benefits. Adding parity bits increases reliability but it also increases the amount of data which is stored or transmitted

In conclusion, I think, it’s essential to understand error detection and correction because everything in computer science depends on precise and accurate data. It doesn’t matter if it’s a small device or huge data center, in every cases those techniques ensure that our systems are secure and trustworthy.
