1. I chose the computational task that finds the maximum value between three inputed numbers.

```python
#include <stdio.h>
#include <iostream>
using namespace std;

int main() {
    int a, b, c;
    cout<<"enter three numbers: ";
    cin>>a>>b>>c;
    int max;
    if (a>=b && a>=c) max = a;
    else if (b>=a && b>=c) max = b;
    else max = c;
    cout <<"Maximal number is: "<<max<<endl;
    return 0;
}
// I couldn't download C++ kernel for jupyter notebook, so it can't run this code here. But I tested it in 
//C++ compiler and it works fine.
```

When I compile my program, the C++ compiler first translates my source code into assembly code (checking for errors along the way), and then the assembler turns it into a binary instruction that the CPU can execute. The linker also adds the standard library parts that my program needs, like iostream.
In this code i only used local variables (int max,a,b,c) which are stored on the stack, because they are simple local integers. The heap is not used at all here, because i didn't allocate anything dynamicly (pointers, etc.).
Operating system is the one that actually launches my code when i run it. It handles the reading of the input and the printing of the result too. It creates a new process for my program and loads it into memory, because apps can only run from memory, Also operating system sets up things like the stack and the heap.
During execution, scheduler inside the operating system decides which program gets CPU next time. CPU is using interrupts, which notifies operating system to for example, read input (cin) or that there's just invalid instructions etc. It tries to arrange everything in a fair way, since many processes may be running at the same time. Also at the end return 0 indicates operating system that there were no errors and program finished successfuly

```python
#include <iostream>
#include <stdio.h>
using namespace std;

int findMax(int a, int b, int c) {
    if (a>=b && a>=c){
        return a;
    }
    else if (b>=a && b>=c){
        return b;
    }
    else {
        return c;
    }
}

int main()
{
    int max, a, b, c;
    cout<<"enter three numbers: ";
    cin>>a>>b>>c;
    max = findMax(a, b, c);
    cout<<"Maximal number is: "<<max<<endl;
    return 0;
}

```

I added function to my code in the previous code shell, which still does the same program (finds max number between inputs).
When we add a new function like, findMax, and then call it from main program, computer temporarily pauses main's work and jumps to a new function. This entire process is managed by stack using stack frames. Everytime we call a new function, new stack frame is pushed onto the stack.

Inside this stack frame, the program stores:

the values of the parameters a, b, and c

the local variable max

and the return address so it knows where to go back in main()

So while the function is running, the stack becomes a little bigger.
When the function finishes and returns the maximum number, its whole stack frame is removed, so the stack shrinks, and the program continues where it left off in main().
